```python
df['Selling_Price'] = df['Selling_Price'] * 100000
df['Present_Price'] = df['Present_Price'] * 100000
```

```python
import datetime
date_time = datetime.datetime.now()
df['Age'] = date_time.year - df['Year']
df.drop('Year',axis = 1 ,inplace=True)
df.head(4)
```

```python
import seaborn as sns
sns.boxplot(df['Selling_Price'])
```

```python
sorted(df['Selling_Price'], reverse=True)
```

```python
 df = df[~(df['Selling_Price'] >= 3300000) & (df['Selling_Price'] <= 3500000)]
```

```python
df.shape
```

```python
df.head(1)
```

```python
df['Fuel_Type'].unique()
```

```python
df['Fuel_Type'] =  df['Fuel_Type'].map({'Petrol' : 0, 'Diesel' : 1, 'CNG' : 2})
df.head(1)
```

```python
df['Seller_Type'].unique()
```

```python
df['Seller_Type'] = df['Seller_Type'].map({'Dealer' : 0, 'Individual' : 1})
df.head(1)
```

```python
df['Transmission'].unique()
```

```python
df['Transmission'] = df['Transmission'].map({'Manual' : 0, 'Automatic' : 1})
df.head(10)
```

```python
X = df.drop(['Car_Name', 'Selling_Price'], axis = 1)
Y = df['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test = train_test_split(X, Y, random_state = 42,test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)

rf = RandomForestRegressor()
rf.fit(x_train, y_train)

gbr = GradientBoostingRegressor()
gbr.fit(x_train, y_train)

xgb = XGBRegressor()
xgb.fit(x_train, y_train)

```

```python
lr_prediction = lr.predict(x_test)
rf_prediction = rf.predict(x_test)
gbr_prediction = gbr.predict(x_test)
xgb_prediction = xgb.predict(x_test)
```

```python
from sklearn.metrics import r2_score

lr_score = r2_score(y_test, lr_prediction)
rf_score = r2_score(y_test, rf_prediction)
gbr_score = r2_score(y_test, gbr_prediction)
xgb_score = r2_score(y_test, xgb_prediction)
```

```python
print('lr_score: ', lr_score)
print('rf_score: ', rf_score)
print('gbr_score: ', gbr_score)
print('xgb_score: ', xgb_score)
```

```python
final_results = pd.DataFrame({'Models' : ['LR', 'RF', 'GBR', 'XGB'],
              'R2_Score' :[ 0.6791, 0.7208, 0.8680, 0.8889]})
```

```python
final_results
```

```python
sns.barplot(x='Models', y='R2_Score', data=final_results)
```

```python
xg = XGBRegressor()
xg_final = xg.fit(X, Y)
```
